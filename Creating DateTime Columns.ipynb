```python
import pandas as pd
groceries = pd.read_excel('Groceries.xlsx')
groceries.head()
```

```python
#creating new column 'Last_Updated_Time'

groceries['Last_Updated_Time'] = groceries.Last_Updated.dt.time 
groceries.head()
```

```python
# creating new column
groceries['Shipment_Date_DOW'] = groceries.Next_Scheduled_Shipment.dt.dayofweek
groceries.head()
```

```python
#converting'Shipment_Date_DOW' from numeric into day
dow_mapping = {0:'Monday',
               1:'Tuesday',
               2:'Wednesday',
               3:'Thursday',
               4:'Friday',
               5:'Saturday',
               6:'Sunday'}
groceries['Shipment_Date_DOW'] = groceries['Shipment_Date_DOW'].map(dow_mapping)
groceries.head()
```

```python
# adding 1 day with column 'Next_Scheduled_Shipment'

groceries['New Scheduled Shipment'] = groceries.Next_Scheduled_Shipment + pd.to_timedelta(1, 'D')
groceries.head()
```

```python
# creating 'Updated Scheduled Shipment' new column if category is 'Produce: Fruit'

import numpy as np
groceries['Updated Scheduled Shipment'] = np.where(groceries.Category == 'Produce: Fruit',
        groceries.Next_Scheduled_Shipment + pd.to_timedelta(1, 'D'),
        groceries.Next_Scheduled_Shipment)

groceries.head(30)
```
