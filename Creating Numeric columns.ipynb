```python
import pandas as pd
groceries = pd.read_excel('Groceries.xlsx')
groceries.head()
```

```python
# by adding 5% tax and rounding 2 digits

round(groceries.Price_Dollars * 1.05, 2)
```

```python
# creating new columns by adding 5% tax and rounding 2 digits

groceries ['New_Price_Dollars'] = round(groceries.Price_Dollars * 1.05, 2)
```

```python
groceries.head()
```

```python
#calculating what % of each inventory of total inventory hold in the warehouse
groceries['Total_Inventory'] = groceries.Inventory.sum()
groceries['Inventory %'] = round(groceries['Inventory'] / groceries['Total_Inventory'] * 100,2)
```

```python
groceries.head()
```

```python
# creating new column to show inventory status

import numpy as np

conditions = [
    groceries.Total_Inventory < 50,
    (groceries.Total_Inventory >= 50) & (groceries.Total_Inventory <= 1000),
    groceries.Total_Inventory > 1000
]

choices = ['Low', 'Medium', 'High']

groceries['Inventory_Status'] = np.select(conditions, choices, default='Unknown')
```
